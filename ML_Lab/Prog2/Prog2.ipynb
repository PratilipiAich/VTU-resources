```python
import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv('enjoysport.csv')
print("The Training Data Set:\n")
concepts = np.array(data.iloc[:,0:-1])
target = np.array(data.iloc[:,-1]) 
print("Concepts:\n", concepts) 
print("\nTarget:\n", target)
```

```python
def learn(concepts, target): 
    S = concepts[0].copy()     
    print("\nInitialization:")     
    print("S: ", S)  
    G = [["?" for i in range(len(S))] for i in range(len(S))]     
    print("G: ", G)  
    print("\n")
    
    for i, h in enumerate(concepts):
        if target[i] == "yes":
            for x in range(len(S)): 
                if h[x]!= S[x]:                    
                    S[x] ='?'                     
                    G[x][x] ='?'
                   
        if target[i] == "no":            
            for x in range(len(S)): 
                if h[x]!= S[x]:                    
                    G[x][x] = S[x]                
                else:                    
                    G[x][x] = '?'        
      
        print("S",i+1, ": ", S)         
        print("G",i+1, ": ", G)
        print("\n")

    indices = [i for i, val in enumerate(G) if val == ['?', '?', '?', '?', '?', '?']]    
    for i in indices:   
        G.remove(['?', '?', '?', '?', '?', '?']) 
    return S, G 
```

```python
s_final, g_final = learn(concepts, target)
print("Final Specific hypothesis:", s_final,"\n")
print("Final General hypothesis:", g_final,"\n")
```
